# If SPY closes within 0.5 points of its high on Friday,
# what happens on the next trading day?

## Get 2 years of daily data for SPY

```python
spy = yf.Ticker("SPY")
daily_ohlc = spy.history(start="2021-12-08", end="2023-12-07", interval="1D")
# Identify the day of the week for each day
daily_ohlc['day_of_week'] = daily_ohlc.index.dayofweek
```

## Identify setups and calculate next day returns

```python
#Calculate the next day's net change
daily_ohlc['next_day_change'] = daily_ohlc['Close'].shift(-1) - daily_ohlc['Close']

#Select Fridays (day_of_week = 4) on which the close is less than 0.5 points below the high
daily_ohlc['setup'] = (daily_ohlc['day_of_week'] == 4) & (daily_ohlc['Close'] >= daily_ohlc['High'] - 0.5)

# Get results and desciptive statistcs
result_df = daily_ohlc.loc[daily_ohlc['setup'] == True]
result_df['next_day_change'].describe()
```
